```python
"""
Student Grade Calculator
Author: [Your Name]
Date: [Current Date]
Description: Calculates grades based on marks with encouraging messages
"""

def calculate_grade(marks):
    """
    Calculate grade based on marks with encouraging messages

    Args:
        marks (int): Student's marks (0-100)

    Returns:
        tuple: (grade, message) containing grade letter and encouraging message
    """
    # Grading logic using if-elif-else statements
    if marks >= 90:
        grade = 'A'
        message = "Outstanding! You've excelled beyond expectations! ðŸŒŸ"
    elif marks >= 80:
        grade = 'B'
        message = "Excellent work! You're doing great! ðŸŽ¯"
    elif marks >= 70:
        grade = 'C'
        message = "Good job! Solid performance! ðŸ‘"
    elif marks >= 60:
        grade = 'D'
        message = "You passed! Keep pushing to improve! ðŸ’ª"
    else:
        grade = 'F'
        message = "Don't give up! Review and try again, you can do it! ðŸ“š"

    return grade, message

def validate_marks(marks):
    """
    Validate that marks are between 0 and 100

    Args:
        marks (str): Input marks as string

    Returns:
        tuple: (is_valid, validated_marks, error_message)
    """
    try:
        marks_int = int(marks)
        if 0 <= marks_int <= 100:
            return True, marks_int, ""
        else:
            return False, None, "Marks must be between 0 and 100!"
    except ValueError:
        return False, None, "Please enter a valid number!"

def display_results(student_name, marks, grade, message):
    """
    Display formatted results

    Args:
        student_name (str): Name of student
        marks (int): Student's marks
        grade (str): Grade letter
        message (str): Encouraging message
    """
    print("\n" + "="*50)
    print("GRADE REPORT".center(50))
    print("="*50)
    print(f"Student Name: {student_name}")
    print(f"Marks Obtained: {marks}/100")
    print(f"Grade: {grade}")
    print(f"Message: {message}")
    print("="*50)

def main():
    """
    Main function to run the grade calculator
    """
    print("="*50)
    print("STUDENT GRADE CALCULATOR".center(50))
    print("="*50)

    # Get student name
    student_name = input("Enter student name: ").strip()

    # Input validation loop
    while True:
        marks_input = input("Enter marks (0-100): ").strip()

        # Validate input
        is_valid, marks, error_msg = validate_marks(marks_input)

        if is_valid:
            # Calculate grade
            grade, message = calculate_grade(marks)

            # Display results
            display_results(student_name, marks, grade, message)
            break
        else:
            print(f"âŒ Error: {error_msg}")
            print("Please try again.\n")

def run_test_cases():
    """
    Run predefined test cases to verify grading logic
    This demonstrates the use of functions for testing
    """
    print("\n" + "="*50)
    print("TESTING GRADE CALCULATOR".center(50))
    print("="*50)

    test_cases = [
        (95, 'A'),
        (85, 'B'),
        (75, 'C'),
        (65, 'D'),
        (55, 'F'),
        (100, 'A'),
        (0, 'F'),
        (90, 'A'),
        (80, 'B'),
        (70, 'C'),
        (60, 'D')
    ]

    print(f"{'Marks':<10} {'Expected Grade':<15} {'Actual Grade':<15} {'Status':<10}")
    print("-"*50)

    for marks, expected_grade in test_cases:
        actual_grade, _ = calculate_grade(marks)
        status = "âœ“ PASS" if actual_grade == expected_grade else "âœ— FAIL"
        print(f"{marks:<10} {expected_grade:<15} {actual_grade:<15} {status:<10}")

# Check if this is the main program being run
if __name__ == "__main__":
    while True:
        main()

        # Ask if user wants to continue or run tests
        print("\nOptions:")
        print("1. Calculate another grade")
        print("2. Run test cases")
        print("3. Exit")

        choice = input("\nEnter your choice (1-3): ").strip()

        if choice == '2':
            run_test_cases()
            continue_option = input("\nPress Enter to continue...")
        elif choice == '3':
            print("\nThank you for using Student Grade Calculator!")
            print("Goodbye! ðŸ‘‹")
            break
        elif choice != '1':
            print("Invalid choice. Continuing with grade calculation...")
        print("\n")
```
